Aggregate Sensor Readings:

    Combine the sensor readings from TP2, TP3, H1, DV_pressure, Reservoirs, and Oil_temperature into a single aggregated feature. This can be done by calculating the mean, median, or standard deviation of these sensor values

```python


# Calculate the mean of sensor readings
new_df['Mean_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].mean(axis=1)

# Calculate the median of sensor readings
new_df['Median_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].median(axis=1)

# Calculate the standard deviation of sensor readings
new_df['StdDev_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].std(axis=1)

# Display the updated dataframe
print(new_df)
```

Sensor Differential:

    Calculate the difference between two sensor readings. For example, you can subtract TP2 from TP3 or calculate the difference between Oil_temperature and DV_pressure. This differential may capture certain relationships.

```python

# Calculate the difference between TP3 and TP2
new_df['TP3_minus_TP2'] = new_df['TP3'] - new_df['TP2']

# Calculate the difference between Oil_temperature and DV_pressure
new_df['OilTemp_minus_DVPressure'] = new_df['Oil_temperature'] - new_df['DV_pressure']

# Display the updated dataframe
print(new_df)
```

Sensor Ratios:

    Create ratios between sensor readings. For instance, you can calculate the ratio of DV_pressure to Oil_temperature. This can be useful if certain sensor ratios have physical significance in your metro train system.

```python


# Calculate the ratio of DV_pressure to Oil_temperature
new_df['Pressure_to_Temperature_Ratio'] = new_df['DV_pressure'] / new_df['Oil_temperature']

# Display the updated dataframe
print(new_df)
```

Analog sensors: • measure the pressure generated in the pneumatic panel (TP3), • measure the current of one phase of the three-phase motor (Motor-Current) Digital sensors: • measure the electrical signal of air intake valve on compressor (COMP) • measure the electrical signal of the solenoid valve (DV Electric) • indicate which tower is drying the air and which tower is draining the humidity removed from the air (TOWERS) • activate the COMP intake valve to start the compressor under load (MPG) • indicate when the pressure is lower than 7 bars (LPS) • detect the discharge in the air drying towers (PressureSwitch) • count the absolute amount of air passing from the APU to the reservoirs (Caudal)

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_regression

# Assuming your dataframe is named 'df'
analog_columns = ['TP3', 'Motor_current']
digital_columns = ['COMP', 'DV_eletric', 'Towers', 'MPG', 'LPS', 'Pressure_switch', 'Caudal_impulses']

# Create dataframes for analog and digital sensors
analog_data = df[analog_columns]
digital_data = df[digital_columns]

# Calculate mutual information between analog and digital sensors
mi_analog_digital = []
for analog_sensor in analog_columns:
    mi_values = []
    for digital_sensor in digital_columns:
        mi = mutual_info_regression(analog_data[analog_sensor].values.reshape(-1, 1), digital_data[digital_sensor])
        mi_values.append(mi)
    mi_analog_digital.append(mi_values)

# Create a DataFrame to store mutual information values
mi_df = pd.DataFrame(mi_analog_digital, columns=digital_columns, index=analog_columns)

# Now mi_df contains the mutual information values between analog and digital sensors
print(mi_df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



# Define X (independent variable) and Y (dependent variable)
X = df[['TP2']]
Y = df['Motor_current']

# Split the data into training, validation, and test sets
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, Y_train)

# Make predictions on the validation set
predictions_val = model.predict(X_val)

# Make predictions on the test set
predictions_test = model.predict(X_test)

# Calculate Mean Squared Errors (MSE) for validation and test sets
mse_val = mean_squared_error(Y_val, predictions_val)
mse_test = mean_squared_error(Y_test, predictions_test)

# Plot the original data points and the linear regression line on the validation set
plt.scatter(X_val, Y_val)
plt.plot(X_val, predictions_val, color='red', label='Linear Regression Line')
plt.xlabel('TP2')
plt.ylabel('Motor_current')
plt.title('Linear Regression Validation Predictions')
plt.legend()
plt.show()

# Plot the original data points and the linear regression line on the test set
plt.scatter(X_test, Y_test, label='Test Data')
plt.plot(X_test, predictions_test, color='green', label='Linear Regression Line')
plt.xlabel('TP2')
plt.ylabel('Motor_current')
plt.title('Linear Regression Test Predictions')
plt.legend()
plt.show()

# Print the coefficients (slope) and intercept
print(f'Coefficient (Slope): {model.coef_[0]}')
print(f'Intercept: {model.intercept_}')

# Print the Mean Squared Errors (MSE) for validation and test sets
print(f'Mean Squared Error (MSE) on training set: {mse_val}')
print(f'Mean Squared Error (MSE) on Test Set: {mse_test}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score,make_scorer

# Assuming 'Timestamp' is a datetime column, you may need to convert it to datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Define the target variable and exogenous variables
target_col = 'Motor_current'
exog_cols = ['TP2', 'DV_eletric', 'Oil_temperature', 'TP3', 'Reservoirs', 'DV_pressure', 'Towers', 'H1', 'COMP', 'MPG']

# Create lag features for the target variable (e.g., lag 1, lag 2)
for lag in range(1, 3):
    df[f'{target_col}_lag{lag}'] = df[target_col].shift(lag)

# Drop rows with missing values in the lag features
df.dropna(subset=[f'{target_col}_lag1', f'{target_col}_lag2'], inplace=True)

# Define train, validation, and test periods
train_start_date = pd.to_datetime('2020-02-01')
val_start_date = pd.to_datetime('2020-04-01')
test_start_date = pd.to_datetime('2020-08-01')

train_data = df[df['timestamp'] < val_start_date]
val_data = df[(df['timestamp'] >= val_start_date) & (df['timestamp'] < test_start_date)]
test_data = df[df['timestamp'] >= test_start_date]

# Define X and Y for train, validation, and test sets
X_train = train_data[exog_cols + [f'{target_col}_lag1', f'{target_col}_lag2']]
y_train = train_data[target_col]

X_val = val_data[exog_cols + [f'{target_col}_lag1', f'{target_col}_lag2']]
y_val = val_data[target_col]

X_test = test_data[exog_cols + [f'{target_col}_lag1', f'{target_col}_lag2']]
y_test = test_data[target_col]

# Define a range of alpha values to test
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

# Create a dictionary of hyperparameters to search
param_grid = {'alpha': alphas}
# Define a custom scoring function that calculates negative mean squared error
def custom_neg_mean_squared_error(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return -mse  # Return negative MSE for minimization

# Create the scoring function using make_scorer
custom_scoring = make_scorer(custom_neg_mean_squared_error)


# Create the grid search object
grid_search = GridSearchCV(Lasso(random_state=42), param_grid, cv=5, scoring=custom_scoring)

# Fit the grid search to your data
grid_search.fit(X_train, y_train)

# Get the best alpha value from the grid search
best_alpha = grid_search.best_params_['alpha']

# Create and train a Lasso regression model with the best alpha
model = Lasso(alpha=best_alpha, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the training set
y_train_pred = model.predict(X_train)

# Calculate evaluation metrics for the training set
train_mse = mean_squared_error(y_train, y_train_pred)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

# Make predictions on the validation set
y_val_pred = model.predict(X_val)

# Calculate evaluation metrics for the validation set
val_mse = mean_squared_error(y_val, y_val_pred)
val_mae = mean_absolute_error(y_val, y_val_pred)
val_r2 = r2_score(y_val, y_val_pred)

# Make predictions on the test set
y_test_pred = model.predict(X_test)

# Calculate evaluation metrics for the test set
test_mse = mean_squared_error(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Print the optimal alpha value
print(f"Optimal Alpha from Grid Search: {best_alpha}")

# Print validation and test metrics
print("Training Set Metrics:")
print(f"Mean Squared Error (MSE): {train_mse:.2f}")
print(f"Mean Absolute Error (MAE): {train_mae:.2f}")
print(f"R-squared (R2): {train_r2:.2f}")

print("Validation Set Metrics:")
print(f"Mean Squared Error (MSE): {val_mse:.2f}")
print(f"Mean Absolute Error (MAE): {val_mae:.2f}")
print(f"R-squared (R2): {val_r2:.2f}")

print("\nTest Set Metrics:")
print(f"Mean Squared Error (MSE): {test_mse:.2f}")
print(f"Mean Absolute Error (MAE): {test_mae:.2f}")
print(f"R-squared (R2): {test_r2:.2f}")
```
