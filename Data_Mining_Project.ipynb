```python
iqr_multiplier = 1.5

# Calculate the IQR for all numeric columns
Q1 = df.select_dtypes(include=['number']).quantile(0.25)
Q3 = df.select_dtypes(include=['number']).quantile(0.75)
IQR = Q3 - Q1

# Create upper and lower bounds to identify outliers
lower_bound = Q1 - (iqr_multiplier * IQR)
upper_bound = Q3 + (iqr_multiplier * IQR)

# Create a boolean mask to identify rows with outliers
outliers = ((df.select_dtypes(include=['number']) < lower_bound) | (df.select_dtypes(include=['number']) > upper_bound)).any(axis=1)

# Remove rows with outliers
df_no_outliers = df[~outliers]
```

To create a new column that may help predict or explain variations in motor current using the columns TP2, TP3, H1, DV_pressure, Reservoirs, and Oil_temperature.

Aggregate Sensor Readings:

    Combine the sensor readings from TP2, TP3, H1, DV_pressure, Reservoirs, and Oil_temperature into a single aggregated feature. This can be done by calculating the mean, median, or standard deviation of these sensor values

```python


# Calculate the mean of sensor readings
new_df['Mean_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].mean(axis=1)

# Calculate the median of sensor readings
new_df['Median_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].median(axis=1)

# Calculate the standard deviation of sensor readings
new_df['StdDev_Sensor'] = new_df[['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature']].std(axis=1)

# Display the updated dataframe
print(new_df)
```

Sensor Differential:

    Calculate the difference between two sensor readings. For example, you can subtract TP2 from TP3 or calculate the difference between Oil_temperature and DV_pressure. This differential may capture certain relationships.

```python

# Calculate the difference between TP3 and TP2
new_df['TP3_minus_TP2'] = new_df['TP3'] - new_df['TP2']

# Calculate the difference between Oil_temperature and DV_pressure
new_df['OilTemp_minus_DVPressure'] = new_df['Oil_temperature'] - new_df['DV_pressure']

# Display the updated dataframe
print(new_df)
```

Sensor Ratios:

    Create ratios between sensor readings. For instance, you can calculate the ratio of DV_pressure to Oil_temperature. This can be useful if certain sensor ratios have physical significance in your metro train system.

```python


# Calculate the ratio of DV_pressure to Oil_temperature
new_df['Pressure_to_Temperature_Ratio'] = new_df['DV_pressure'] / new_df['Oil_temperature']

# Display the updated dataframe
print(new_df)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Standardize the features (mean=0, std=1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(new_df)

# Initialize PCA with the desired number of components
num_components = 5  # Choose the number of components you want to retain
pca = PCA(n_components=num_components)

# Apply PCA to the standardized data
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components as columns
pca_new_df = pd.DataFrame(data=pca_result, columns=[f'PC{i+1}' for i in range(num_components)])

# Display the PCA-transformed data
print(pca_new_df.head())
```

Mutual Information:

    Calculate the mutual information between each feature and the target variable. Features with higher mutual information are considered more informative.

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_regression



# Separate features (X) and target variable (y)
X = new_df.drop(columns=['Motor_current'])  # Exclude the target variable
y = new_df['Motor_current']

# Calculate mutual information for each feature
mutual_info_scores = mutual_info_regression(X, y)

# Create a DataFrame to store feature names and their mutual information scores
mutual_info_new_df = pd.DataFrame({'Feature': X.columns, 'Mutual_Information': mutual_info_scores})

# Sort features by mutual information score (descending order)
mutual_info_new_df = mutual_info_new_df.sort_values(by='Mutual_Information', ascending=False)

# Display the sorted mutual information scores
print(mutual_info_new_df)
```
